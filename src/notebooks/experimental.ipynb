```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import random
import os
```

```python
# Experimenting with a uniform distribution vs beta distribution for risk 
# Uniform is a naive assumption, Beta is more accurate for banks
uniform_score = random.uniform(0, 0.2) 
beta_score = random.betavariate(1, 100)
```

```python
random.seed(42)
sample_size = 10000

uniform_data = [random.uniform(0, 0.2) for _ in range(sample_size)]
beta_data = [random.betavariate(1, 100) for _ in range(sample_size)]

plt.figure(figsize=(8, 4))

# Uniform distribution
plt.subplot(1, 2, 1)
sns.histplot(uniform_data, bins=50, color='skyblue', kde=True)
plt.title('Uniform Distribution (0.0 to 0.2)"')
plt.xlabel('Risk Score')
plt.ylabel('Count')

# Beta distribution
plt.subplot(1, 2, 2)
sns.histplot(beta_data, bins=50, color='salmon', kde=True)
plt.title('Beta Distribution (Alpha=1, Beta=100)')
plt.xlabel('Risk Score')
plt.xlim(0, 0.2)
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
